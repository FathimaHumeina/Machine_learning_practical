```python
import pandas as pd
```

```python
carPrice=pd.read_csv('/content/drive/MyDrive/DSPractical/carprices.csv')
```

```python
carPrice
```

```python
dummies=pd.get_dummies(carPrice['Car Model'])
```

```python
merged=pd.concat([carPrice,dummies],axis='columns')
```

```python
final_df=merged.drop(['Car Model','Audi A5'] , axis='columns')
```

```python
final_df
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
x=final_df.drop('Sell Price($)',axis='columns')
```

```python
y=final_df['Sell Price($)']
```

```python
model.fit(x,y)
```

```python
model.predict([[45000,4,0,1]])
```

```python
model.predict([[86000,7,1,0]])
```

```python
model.score(x,y)
```

**Pre Processing One Hot Encoding**

```python
dfle=carPrice
```

```python
dfle
```

```python
dfle.dtypes
```

```python
dfle['Car Model'].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder()
```

```python
feature_array=ohe.fit_transform(dfle[['Car Model']]).toarray()
```

```python
feature_labels=ohe.categories_
```

```python
import numpy as np
```

```python
np.array([feature_labels]).ravel()
```

```python
df=pd.DataFrame(feature_array,columns=feature_labels)
```

```python
df
```

```python
final_df=pd.concat([df,dfle],axis=1)
```

```python
final_df
```

```python
final_df.drop(['Car Model'],axis=1,inplace=True)
```

```python
final_df
```

```python
x=final_df.drop(['Sell Price($)'],axis=1).values
y=final_df['Sell Price($)'].values
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.predict([[0,0,1,45000,4]])
```

```python
model.predict([[0,1,0,86000,7]])
```

```python
model.score(x,y)
```
