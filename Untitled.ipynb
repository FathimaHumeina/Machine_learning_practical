```python
heart7=heart5.copy()
```

```python
heart7.ChestPainType.replace({
    'ATA':1,
    'NAP':2,
    'ASY':3,
    'TA':4
},inplace=True)

heart7.RestingECG.replace({
    'Normal':1,
    'ST':2,
    'LVH':3
},inplace=True)

heart7.ExerciseAngina.replace({
    'N':0,
    'Y':1
},inplace=True)

heart7.ST_Slope.replace({
    'Up':1,
    'Flat':2,
    'Down':3
},inplace=True)
```

```python
heart_final=pd.get_dummies(heart7,drop_first=True)
```

```python
heart_final
```

```python
x=heart_final.drop('HeartDisease',axis='columns')
y=heart_final.HeartDisease
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x_scaled,y,random_state=10)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
```

```python
score=cross_val_score(DecisionTreeClassifier(),x,y,cv=5)
```

```python
score.mean()
```

```python
from sklearn.ensemble import BaggingClassifier
bag_model=BaggingClassifier(
base_estimator=DecisionTreeClassifier(),
n_estimators=100,
max_samples=0.8,
oob_score=True,
random_state=0)
```

```python
bag_model.fit(X_train,y_train)
```

```python
bag_model.oob_score_
```

```python
bag_model.score(X_test,y_test)
```

```python
from sklearn.svm import SVC
```

```python
score=cross_val_score(SVC(),x,y,cv=5)
```

```python
score.mean()
```

```python
bag_model=BaggingClassifier(
base_estimator=SVC(),
n_estimators=100,
max_samples=0.8,
oob_score=True,
random_state=0)
```

```python
bag_model.fit(X_train,y_train)
```
